## Missing Values

```python
#Get Total and % of Missings
def missingcheck(df):
 total = df.isnull().sum().sort_values(ascending=False)
 percent_1 = df.isnull().sum()/df.isnull().count()*100
 percent_2 = (np.round(percent_1, 1)).sort_values(ascending=False)
 missing_data = pd.concat([total, percent_2], axis=1, keys=["Total", "%"])
 return missing_data
```

```python
missingcheck(df)
```

### Approach 3: Iterative Imputation 
#### Dá valores negativos em ratio variables -> não usar 

Iterative imputation refers to a process where each feature is modeled as a function of the other features. In our case, BayesianRidge method was applied, which is a regularized linear regression

```python
numerical_vars = df.iloc[:,2:21]
num_vars = numerical_vars.drop(columns=['Regional_indicator', 'Code'])
```

```python
# define imputer
imputer_ii = IterativeImputer(estimator= BayesianRidge(), n_nearest_features=None, imputation_order='ascending')
# fit on the dataset
imputer_ii.fit(num_vars)
# transform the dataset
imputer_ii.transform(num_vars)
num_vars[:]= imputer_ii.transform(num_vars)
iidf=num_vars[:]
print(iidf)
```

```python
iidf.describe()
```

```python
shadyrows3 = iidf[iidf['Marriages']<0]
print(shadyrows3)
```
